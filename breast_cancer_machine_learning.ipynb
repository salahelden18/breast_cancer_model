```python
data = pd.read_csv("data/data.csv")
```

```python
# drop id (unimportant factor) and drop unnamed: 32
data.drop("id", axis=1, inplace=True)
data.drop("Unnamed: 32", axis=1, inplace=True)
```

```python
# according the the graphs below we see that we have some outliers

fig, ((ax0, ax1), (ax2, ax3)) = plt.subplots(figsize=(10,6), nrows= 2, ncols=2)

ax0.hist(data["radius_mean"], bins=20)
ax1.hist(data["texture_mean"], bins=20)
ax2.hist(data["perimeter_mean"], bins=20)
ax3.hist(data["area_mean"], bins=20);
```

```python
data.describe()
```

```python
# drop the missing values in diagnosis and show how much missing values we have
data.isna().sum()
# drop the row where there is missing data in diagnosis Column
data.dropna(subset=["diagnosis"], inplace=True)
```

```python
#Splitting the data
np.random.seed(42)

from sklearn.model_selection import train_test_split

X = data.drop("diagnosis", axis=1)
y = data["diagnosis"]

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# outliers
from sklearn.neighbors import LocalOutlierFactor

clf = LocalOutlierFactor(n_neighbors=20)

outliers = clf.fit_predict(X)

count =0

for i in outliers:
    if(i == -1):
        count = count + 1

count
```

```python
# Filling missing values in the independent features

from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer

mean_imputer = SimpleImputer(strategy="mean")

features = ["radius_mean",
            "texture_mean",
            "perimeter_mean",
            "area_mean",
            "smoothness_mean",
            "compactness_mean",
            "concavity_mean",
            "concave points_mean",
            "symmetry_mean",
            "fractal_dimension_mean",
            "radius_se",
            "texture_se",
            "perimeter_se",
            "area_se",
            "smoothness_se",
            "compactness_se",
            "concavity_se",
            "concave points_se",
            "symmetry_se",
            "fractal_dimension_se",
            "radius_worst",
            "texture_worst",
            "perimeter_worst",
            "area_worst",
            "smoothness_worst",
            "compactness_worst",
            "concavity_worst",
            "concave points_worst",
            "symmetry_worst",
            "fractal_dimension_worst"
           ]

imputer = ColumnTransformer([("mean_imputer", mean_imputer, features)])

filled_x_train = imputer.fit_transform(x_train)
filled_x_test = imputer.transform(x_test)
```

```python
pd.DataFrame(filled_x_train)
```

```python
# Turning the categorical data into numerical (we just have the diagnosis)
from sklearn.preprocessing import LabelEncoder

trasform = LabelEncoder()

transformed_y_train = trasform.fit_transform(y_train)
transformed_y_test = trasform.transform(y_test)
```

```python
len(transformed_y_train), len(transformed_y_test)
```

```python
# implement imbalanced data.
y_data = np.concatenate((transformed_y_train, transformed_y_test))

y_dataframe = pd.DataFrame(y_data)
# the 60% of the data is 0 and 40% of the data is 1 so no need to implement the imbalanced data step.
y_dataframe.value_counts()
```

```python
y_dataframe.plot.hist();
```

```python
# doing the Normalization using (z-score normalization)

from sklearn.preprocessing import StandardScaler

scale = StandardScaler()

scaled_x_train = scale.fit_transform(filled_x_train)
scaled_x_test = scale.transform(filled_x_test)
```

```python
# fitting the model
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier()

clf.fit(scaled_x_train, transformed_y_train)
```

```python
clf.score(scaled_x_test, transformed_y_test)
```

```python
# Another Evalution metric
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier()

cross_val_score = cross_val_score(clf, X, y)



print(f"The Cross-Validated Accuracy: {np.mean(cross_val_score) * 100:.2f}%")
```
