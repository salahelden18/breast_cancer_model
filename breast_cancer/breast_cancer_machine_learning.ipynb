```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as ply
```

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
df = pd.DataFrame(cancer["data"], columns=cancer["feature_names"])
# target = pd.DataFrame(cancer["target"])
df["target"]  = cancer["target"]
```

```python
data = pd.read_csv("data/data.csv")
```

```python
data
```

```python
# drop id (unimportant factor) and drop unnamed: 32
data.drop("id", axis=1, inplace=True)
data.drop("Unnamed: 32", axis=1, inplace=True)
```

```python
# drop the missing values in diagnosis and show how much missing values we have
data.isna().sum()
data.dropna(subset=["diagnosis"], inplace=True)
```

```python
#Splitting the data
from sklearn.model_selection import train_test_split

X = data.drop("diagnosis", axis=1)
y = data["diagnosis"]

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Filling missing values

from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer

mean_imputer = SimpleImputer(strategy="mean")

features = ["radius_mean",
            "texture_mean",
            "perimeter_mean",
            "area_mean",
            "smoothness_mean",
            "compactness_mean",
            "concavity_mean",
            "concave points_mean",
            "symmetry_mean",
            "fractal_dimension_mean",
            "radius_se",
            "texture_se",
            "perimeter_se",
            "area_se",
            "smoothness_se",
            "compactness_se",
            "concavity_se",
            "concave points_se",
            "symmetry_se",
            "fractal_dimension_se",
            "radius_worst",
            "texture_worst",
            "perimeter_worst",
            "area_worst",
            "smoothness_worst",
            "compactness_worst",
            "concavity_worst",
            "concave points_worst",
            "symmetry_worst",
            "fractal_dimension_worst"
           ]
imputer = ColumnTransformer([("mean_imputer", mean_imputer, features)])

filled_x_train = imputer.fit_transform(x_train)
filled_x_test = imputer.transform(x_test)
```

```python
# Turning the categorical data into numerical (we just have the diagnosis)

from sklearn.preprocessing import LabelEncoder

trasform = LabelEncoder()

transformed_y_train = trasform.fit_transform(y_train)
transformed_y_test = trasform.transform(y_test)
```

```python
# doing the Normalization using (z-score normalization)

from sklearn.preprocessing import StandardScaler

scale = StandardScaler()

scaled_x_train = scale.fit_transform(filled_x_train)
scaled_x_test = scale.transform(filled_x_test)
```

```python
# fitting the model
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier()

clf.fit(scaled_x_train, transformed_y_train)
```

```python
clf.score(scaled_x_test, transformed_y_test)
```
